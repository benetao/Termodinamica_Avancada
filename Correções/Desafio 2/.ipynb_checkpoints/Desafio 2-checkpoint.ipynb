## Definindo os valores iniciais

Nessa seção, definimos as condições iniciais da mistura, onde nós obtemos e/ou definimos os valores com as seguintes unidades:

- Massa (m): gramas
- Volume (V): mL
- Pressão (P): kPa
- Número de mols (Nm): mols
- Massa molar (Mm): gramas/mol
- Fração molar (X): adimensional

Além disso, usamos algumas abreviações relacionadas às unidades para dizer que aquela grandeza é daquele composto, c para ciclohexano, t para tolueno e a ausência faz referência à mistura. Exemplo:

- Xc: fração molar do ciclohexano
- Nmt: número de mols do tolueno
- MassaTotal: massa total da mistura

Por fim, para as grandezas definidas no início e no fim, usamos 0 e f. Exemplo:

- V0: volume inicial da mistura
- Vf: volume final da mistura obtida após a destilação

<font color='green'>Muito bem organizado!</font>

De início, definimos que a massa total seria de uma tonelada, ou seja, 1000000 gramas, e que nossa mistura teria frações molares iguais, ou seja, de 0.5 para cada compostos.

```python
MassaTotal = 1000000 #gramas
Xc0 = 0.5 #Fração molar inicial do ciclohexano
Xt0 = 0.5 #Fração molar inicial do tolueno
```

Com isso, precisamos saber quanto de massa cada composto tem, o que pode ser calculado com a seguinte equação:

$$MassaTotal = Mmc \cdot Xc \cdot NmT + Mmt \cdot Xt \cdot NmT$$

Em poucas palavras: pela soma dos produtos da massa molar pela fração molar de cada composto pelo número mols total.

```python
Mmt = 92.14 #g/mol, massa molar
Mmc = 84.16 #g/mol, massa molar
NmT = 11344.3 #número de mols total
massa_aproximada = Mmc*Xc0*(NmT) + Mmt*Xt0*(NmT)
print(f'O valor aproximado de massa inicial é de cerca de {round(massa_aproximada, 3)} gramas')
```

<font color='yellow'>Para calcular o número total de mols bastava remanejar a equação da seguinte forma:</font>

```python
NmT = MassaTotal/(Mmc*Xc0 + Mmt*Xt0)

print(NmT)
```

Assim, encontramos que o valor de número de mols total para termos uma massa próxima da que definimos é: $NmT = 11344.3$ mols e como a fração molar é 0.5, os valores para número de mols de cada composto são $Nmc = Nmt = 5672.15$ mols.

```python
Nmc0 = 5672.15 # número de mols inicial do ciclohexano
Nmt0 = 5672.15 # número de mols inicial do tolueno
```

Com isso, podemos calcular a massa inicial de cada composto multiplicando o número de mols pela massa molar de cada composto.

$$M = Mm \cdot Nm$$

```python
Mc0 = Mmc*Nmc0 # Massa total inicial do ciclohexano em gramas
Mt0 = Mmt*Nmt0 # Massa total inicial do tolueno em gramas

print(f"A massa inicial de ciclohexano é de {Mc0} gramas")
print(f"A massa inicial de tolueno é de {Mt0} gramas")
```

## Equação de Antoine

Nessa seção, definimos as funções que são usadas para calcular a pressão de vapor dos compostos puros num dado intervalo de temperatura. Então, definimos três funções, uma com a Equação de Antoine e outras duas com os coeficientes de cada substância obtidos experimentalmente e acessados no NIST.

NIST (National Institue of Standards and Technology): https://www.nist.gov/.

A Equação de Antoine é uma fórmula semi-empírica obtida experimentalmente pelo engenheiro francês Louis Charles Antoine em 1888 e pode ser expressa assim:

$$\log{(P)} = A - \frac{B}{C + T}$$

Essa pode ser facilmente modificada para uma forma mais útil e interessante de aplicar:

$$P = 10^{\left(A - \frac{B}{C + T}\right)}$$

```python
def Antoine(T,A,B,C):
    """Essa função fornece o valor de pressão de vapor para as substâncias
    
    Args:
        T: temperatura em Kelvin
        A, B, C: coeficientes calculados para cada substância
        
    Return:
        P: pressão de vapor em kPa
    """
    P = 10**(A - (B/(C+T)))
    return P*100 # Pressão em kPa
```

Coeficientes do tolueno: https://webbook.nist.gov/cgi/cbook.cgi?ID=C108883&Mask=4&Type=ANTOINE&Plot=on

```python
def Antoine_Coef_Tolueno(T):
    """Essa função fornece os coeficientes indicados para o cálculo da pressão de vapor com a equação de Antoine em cada intervalo
    de temperatura.
    
    Args:
        T: temperatura em Kelvin
    
    Return:
        A, B, C: coeficientes para o cálculo da pressão de vapor com a equação de Antoine
    """
    
    if T >= 273.14 and T <= 297.89:
        A= 4.23679
        B= 1426.448
        C= -45.957
        return A, B, C
    
        
    if T >= 297.90 and T <= 303.00:
        A= 4.14157
        B= 1377.578
        C= -50.507
        return A, B, C
    
        
    if T >= 303.01 and T <= 343.00:
        A= 4.08245
        B= 1346.382
        C= -53.508
        return A, B, C
    
        
    if T >= 343.01 and T <= 462.66:
        A= 4.07827
        B= 1343.943
        C= -53.773
        return A, B, C
    
        
    if T >= 462.67 and T <= 580.01:
        A= 4.54436
        B= 1738.123
        C= 0.394
        return A, B, C
```

Coeficintes do ciclohexano: https://webbook.nist.gov/cgi/cbook.cgi?ID=C108883&Mask=4&Type=ANTOINE&Plot=on.

```python
def Antoine_Coef_Ciclohexano(T):
    """Essa função fornece os coeficientes indicados para o cálculo da pressão de vapor com a equação de Antoine em cada intervalo
    de temperatura.
    
    Args:
        T: temperatura em Kelvin
    
    Return:
        A, B, C: coeficientes para o cálculo da pressão de vapor com a equação de Antoine
    """
    
    if T>=293.06 and T<=315.69:
        A= 3.96988
        B= 1203.526
        C= -50.287
        return A, B, C
    
    if T>=315.70 and T<=353.90:
        A= 3.17125
        B= 780.637
        C= -107.29
        return A, B, C
    
    if T>=353.91 and T<=523.01:
        A= 4.13983
        B= 1316.554
        C= -35.581
        return A, B, C
```

## Destilação

Nessa seção, usamos os dados do diagrama de fases atualizado e da pressão de vapor calculada com a Equação de Antoine para saber como deveria ser o aparato necessário para destilar nossa mistura e separar um dos compostos presentes.

Então, criamos a seguinte função para automatizar o processo de destilação. Ela usa da Lei de Dawton e da Lei de Raoult, como expostas abaixo:

- Lei de Dawton: a pressão total em um sistema fechado é igual a soma da pressão parcial de cada componente gasoso.
    $$P_{total} = P_A + P_B$$
- Lei de Raoult: a pressão de vapor de um composto em uma mistura é igual ao produto de sua fração molar pela pressão de vapor no seu estado puro.
    $$P_A = P_{A}^{o} \cdot X_A$$
    
Combinando as duas, temos:
    $$P_{total} = P_{A}^{o} \cdot X_A + P_{B}^{o} \cdot X_B$$

O procedimento descrito pode ser encontrado em mais detalhes no LibreTexts Chemistry: https://chem.libretexts.org/Courses/SUNY_Oneonta/Chem_221%3A_Organic_Chemistry_I_(Bennett)/2%3ALab_Textbook_(Nichols)/05%3A_Distillation/5.02%3A_Simple_Distillation/5.2B%3A_Separation_Theory

Com isso, escolhemos destilar o ciclohexano por possuir o ponto de vaporização mais baixo e maior Entalpia de Combustão ($\Delta H_{combustão}^{ciclohexano} = -3919.9$ kj/mol e $\Delta H_{combustão}^{tolueno} = -3909.9$ kJ/mol).

Valores obtidos novamente no NIST:
- Entalpia de Combustão do Tolueno: https://webbook.nist.gov/cgi/cbook.cgi?ID=C108883&Type=HCOMBL
- Entalpia de Combustão do Ciclohexano: https://webbook.nist.gov/cgi/cbook.cgi?ID=C110827&Type=HCOMBL

<font color='yellow'>O ponto de vaporização mais baixo do ciclohexano é uma condição para a destilação.</font>
