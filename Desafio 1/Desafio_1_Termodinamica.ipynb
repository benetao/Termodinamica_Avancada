### Problema A

Portanto, existe uma relação direta entre a *quantidade de calor* gerada pelo combustível e a *energia* gerada pelo ciclo de rankine.

Sendo assim, é positiva a utilização de <u>**hidrogênio**</u> como combustível, haja vista o seu alto calor de combustão por massa, dessa forma, a partir da combustão de uma pequena massa de hidrogênio, é possível gerar grandes quantidades de calor. 
    
Vale apontar que uma das formas mais estudadas de obtenção de hidrogênio é a partir da separação (quebra) da molécula de <u>**água**</u>, que é um composto muito mais simples de armazenar, levando em consideração que a mistura escolhida no desafio deve estar, inicialmente, no estado líquido.
    
Paralelamente, o <u>**etanol**</u> também se caracteriza como um composto com alta entalpia liberada pelo processo de combustão.  Ademais, a mistura de água com etanol já foi amplamente estudada e caracterizada devido ao seu importante papel no nosso dia a dia. Afinal, tal mistura é muito utilizada não apenas dentro  dos laboratórios, na forma de etanol 70% como reagente ou higienizador, mas também em automóveis, sendo uma das misturas mais usadas como combustível no mundo, juntamente com gasolina+ etanol. Sendo assim, o fácil acesso e disponibilidade de características ( entalpia e diagrama de fase) dessa mistura também foi um aspecto importante para a nossa escolha.

### Problema B


As entalpias de combustão do Hidrogênio e do Etanol utilizadas foram retiradas do site [Engineering Tool Box](https://www.engineeringtoolbox.com/standard-heat-of-combustion-energy-content-d_1987.html) (clique para acessar o link do site). Como podemos analisar na tabela, o gás hidrogênio é, de fato, o que possui maior entalpia de combustão por unidade de massa, o que justifica nossa escolha da água (que será posteriormente hidrolisada para obtenção de hidrogênio) para a mistura.

Ademais, o etanol também apresenta uma das mais altas entalpias de combustão da tabela, o que indica que ele será muito eficiente para produção de energia no ciclo de Rankine.



Para além da entalpia de combustão, a qual justifica o uso dos combustíveis escolhidos, é interessante estabelecer qual é a entalpia de formação de cada componente da mistura individualmente, a diferença de entalpia atribuida ao processo de mistura e, então, estabelecer qual é a entalpia de formação da mistura.

Portanto, foi utilizado o site da [Universidade de Ohio](https://www.ohio.edu/mechanical/thermo/property_tables/combustion/Enth_Formation.html) para estabelecer os valores de entalpia de formação da <u>**água**</u> e do <u>**etanol**</u>.

Outro valor de entalpia que também deve ser considerado é o de *entalpia de mistura*. Esse valor diz respeito à quantidade de energia liberada ao se misturar determinados componentes, e varia conforme a concentração da mistura, a temperatura e a pressão. Considerando condições normais de temperatura e pressão, encontra-se no banco de dados [Dortmund Data Bank](http://www.ddbst.com/ddb-he.html) o seguinte gráfico de entalpia da mistura etanol + água, variando de acordo com a fração molar do álcool etílico.

Por conseguinte, com os dados de entalpia de formação de cada componente e de entalpia da mistura, podemos calcular a entalpia de formação da mistura através da função implementada abaixo:

```python
# valores encontrados
ENTALPIA_AGUA = -285.820*1000 # J/mol
ENTALPIA_ETANOL = -277.690*1000 # J/mol
ENTALPIA_MISTURA_AGUA_ETANOL = -785 # J/mol; com etanol a fração molar de 0.17 (concentração escolhida)
FRACAO_MOLAR_ETANOL = 0.17

def entalpia_de_form_mistura(frac_molar1, entalpia1, entalpia2, entalpia_mix):
    frac_molar2 = 1 - frac_molar1
    return entalpia1*frac_molar1 + entalpia2*frac_molar2 + entalpia_mix

ENTALPIA_AGUA_ETANOL = entalpia_de_form_mistura(FRACAO_MOLAR_ETANOL, ENTALPIA_ETANOL, ENTALPIA_AGUA, ENTALPIA_MISTURA_AGUA_ETANOL)

print("A Entalpia de Formação da Água + Etanol foi calculada como", round(ENTALPIA_AGUA_ETANOL, 2), "J/mol")
```

### Problema C

Os dados relativos ao diagrama de fases da mistura entre Etanol e Água foram obtidos a partir de resultados experimentais divulgados no site [Home Distiller](https://homedistiller.org/forum/viewtopic.php?f=6&t=75697) (clique para acessar).

Os dados experimentais estão contidos na planilha "Diagrama de Fases - Etanol e Água" e foram lidos em Python utilizando a função "read_excel" da biblioteca pandas, para, então, ser plotado o gráfico utilizando a biblioteca matplotlib.

```python
ARQUIVO_EXCEL = "Diagrama de Fases - Etanol e Água.xlsx" # Chamando o arquivo excel com dados experimentais obtido no site Home Distiller
NOME_DA_PLANILHA_O = "Sheet1" # Definindo a planila do arquivo xlsx que será utilizada
LINHA_CABECALHO = 1  # Definindo quantas linhas de cabeçalho existem, considerando que a primeira linha é a número 0
COLUNAS_CONTENDO_OS_DADOS = [1, 4, 7] # Definindo quais colunas utilizaremos para montar o data frame, considerando que a primeira coluna é a número 0

df_et_h2o = pd.read_excel( # Criando o Data Frame do diagrama de fases da mistura escolhida (etanol + água)
    ARQUIVO_EXCEL,
    sheet_name=NOME_DA_PLANILHA_O,
    header=LINHA_CABECALHO,
    usecols=COLUNAS_CONTENDO_OS_DADOS,
)

df_et_h2o.head()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
df_et_h2o.plot('Mole Fraction','C',legend=True, label = "Temperatura de Ebulição", ax=ax1, kind = 'line', color = 'black') # Plotando Temperatura de ebulição
df_et_h2o.plot('Mole Fraction.1','C',legend=True, label = "Composição do Vapor", ax=ax1, kind = 'line', color = 'gray') # Plotando composição do vapor
plt.title('Diagrama de Fases - Etanol e Água') # Definindo título do gráfico
plt.xlabel('Fração Molar de Etanol') # Definindo legendas dos eixos
plt.ylabel('Temperatura (°C)')
plt.show() # Plotando o diagrama de fases da mistura escolhida (etanol + água)
```
