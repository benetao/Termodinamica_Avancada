### Problema A

Como descrito no diagrama de apresentação do problema, será necessário escolher uma mistura, de forma que a combustão de algum dos compostos dessa mistura libere calor, o qual será utilizado para gerar trabalho no ciclo de rankine. Portanto, existe uma relação direta entre a quantidade de calor gerada pelo combustível e a energia gerada pelo ciclo de rankine. Dessa forma, é interessante trabalhar com combustíveis que liberam uma quantidade grande de calor em sua combustão, ou seja, combustíveis que possuem uma grande entalpia de combustão. Sendo assim, é positiva a utilização de hidrogênio como combustível, haja vista o seu alto calor de combustão por massa, dessa forma, a partir da combustão de uma pequena massa de hidrogênio, é possível gerar grandes quantidades de calor. Paralelamente, o etanol também se caracteriza como um composto com alta entalpia liberada pelo processo de combustão. Devido às características dos compostos apresentados, foi escolhida a mistura de etanol e água, pois, a partir dessa, é possível a separação entre o etanol e a água e, posteriormente, a divisão da água em hidrogênio e oxigênio. Dessa forma, seria possível a formação dos dois combustíveis desejados: Hidrogênio e Etanol.

### Problema C

Os dados relativos ao diagrama de fases da mistura entre Etanol e Água foram obtidos a partir de resultados experimentais divulgados em: https://homedistiller.org/forum/viewtopic.php?f=6&t=75697

Os dados experimentais estão contidos na planilha "Diagrama de Fases - Etanol e Água" e foram lidos em Python utilizando a função "read_excel" da biblioteca pandas, para, então, ser plotado o gráfico utilizando a biblioteca matplotlib.

```python
ARQUIVO_EXCEL = "Diagrama de Fases - Etanol e Água.xlsx"
NOME_DA_PLANILHA_O = "Sheet1"
LINHA_CABECALHO = 1  # considerando que a 1a linha é a número 0
COLUNAS_CONTENDO_OS_DADOS = [1, 4, 7] # considerando que a 1a coluna é a número 0

df_et_h2o = pd.read_excel(
    ARQUIVO_EXCEL,
    sheet_name=NOME_DA_PLANILHA_O,
    header=LINHA_CABECALHO,
    usecols=COLUNAS_CONTENDO_OS_DADOS,
)

df_et_h2o.head()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
df_et_h2o.plot('Mole Fraction','C',legend=True, label = "Temperatura de Ebulição", ax=ax1, kind = 'line', color = 'black')
df_et_h2o.plot('Mole Fraction.1','C',legend=True, label = "Composição do Vapor", ax=ax1, kind = 'line', color = 'gray')
plt.title('Diagrama de Fases - Etanol e Água')
plt.xlabel('Fração Molar de Etanol')
plt.ylabel('Temperatura (°C)')
#plt.savefig('Gols e Público.png', dpi=100, format='png', facecolor='white')
plt.show()
```
